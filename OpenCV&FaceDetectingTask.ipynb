```python
model = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
```

```python
faces = model.detectMultiScale(img)
faces
```

```python
faces.shape
```

```python
# We can iterate all the faces  
# for face in faces:
# We can use this 4 coordinate to plot a any rectangle
# x,y,w,h = face
# We want to plot a rectangle
# save = cv2.reactangle(img, (two coordinate of first point), (Bottom right point), (RGB format of a colour), the thickness of rectangle)
# RGB format for Black colour(0,0,0), White colour(25,255,255), Red colour(0,0,255)
```

```python
for face in faces:
  x,y,w,h = face
  img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,0), 6)
```

```python
# face datected
plt.imshow(img)
```
